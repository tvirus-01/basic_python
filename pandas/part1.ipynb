# perform merge oparation

```python
student = ['rahim','korim','jobbar','lyla','sony']
score = ['43','76','87','45','79']
subject = ['math','english']
df1 = pd.DataFrame({'student':student, 'subject':subject[0], 'score':score})
df1
```

```python
student = ['rahim','mokles','kuddus','lyla']
total_score = ['54','86','57','83']
subject = ['english']
df2 = pd.DataFrame({'student':student, 'subject':subject[0], 'score':total_score})
df2
```

```python
#inner merge
df1.merge(df2, on='student', how='inner')
#by default its perform an inner merge
#df1.merge(df2)
```

```python
#left merge
df1.merge(df2, on='student', how='left')
```

```python
#right merge
df1.merge(df2, on='student', how='right')
```

```python
#outer merge
df1.merge(df2, on='student', how='outer')
```

# How to perform Join oparation

```python
player = ['player1','player2','player3']
point = [8,9,6]
game = ['game1','game2','game3']
df3 = pd.DataFrame({'player':player,'points':point,'game':game}, index=['l1','l2','l3'])
df3
```

```python
player = ['player1','player5','player6']
power = ['kick','elbow','stamina']
game = ['game1','game4','game5']
df4 = pd.DataFrame({'players':player,'power':power,'games':game}, index=['l1','l4','l5'])
df4
```

```python
#inner join
df3.join(df4, how='inner')
```

```python
#left join
df3.join(df4, how='left')
```

```python
#right join
df3.join(df4, how='right')
```

```python
#outer join
df3.join(df4, how='outer')
```

# concatenate dataframe

```python
pd.concat([df3,df4], sort=False)
```
